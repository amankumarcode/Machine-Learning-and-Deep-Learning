# Analyzing and Visualizing Missing Values

```python
# get the number of missing data points per column
missing_values_count = dataCopy.isnull().sum()

# look at the # of missing points in the first ten columns
missing_values_count[0:]
```

```python
#missingno is a python library for nice visulaisation of missing number in the data
import missingno as msno

# Nullity or missing values by columns
msno.matrix(df=dataCopy.iloc[:,4:], figsize=(15, 6))
```

# Cleaning and Imputing Values

```python
# DROPPING ROWS WITH LOT OF NAN VALUES
dataCopyDropRowWithThres = dataCopy.copy()
dataCopyDropRowWithThres.dropna(thresh=8,inplace=True)
msno.matrix(df=dataCopyDropRowWithThres.iloc[:,4:], figsize=(12, 6), color=(0.42, 0.1, 0.05))
```

```python
# IMPUTING VALUES IN FIRST CLASS AVERAGE WITH MEAN AND IMPUTING OUTLIER NAN WITH ZERO.
dataCopyFillNAMean = dataCopyDropRowWithThres.copy()
dataCopyFillNAMean["first_class.average"] = pd.to_numeric(dataCopyFillNAMean["first_class.average"])
dataInd = np.mean(dataCopyFillNAMean.teams == 'India,')
dataCopyFillNAMean.iloc[:,5:] = dataCopyFillNAMean.fillna(0)
dataCopyFillNAMean[dataCopyFillNAMean.teams == 'India,']= dataCopyFillNAMean.fillna(dataInd)
dataRest = np.mean(dataCopyFillNAMean.teams != 'India,')
dataCopyFillNAMean[dataCopyFillNAMean.teams != 'India,'] = dataCopyFillNAMean.fillna(dataRest)
msno.matrix(df=dataCopyFillNAMean.iloc[:,4:], figsize=(12, 6), color=(0.42, 0.1, 0.05))
```

# Using Z score Normalisation

```python
data1 = dataCopyFillNAMean.copy()
for i in range(4,18):
    data1.iloc[:,i] = pd.to_numeric(data1.iloc[:,i])
data1.iloc[:,4:] = (data1.iloc[:,4:] - data1.iloc[:,4:].mean())/data1.iloc[:,4:].std()
data1
```

# Using Random Forest for Feature Selection

```python
cric_features1 = data1.iloc[:,4:]
cric_classes1 = data1.teams == 'India'
cricX1 = np.array(cric_features1)
cricY1 = np.array(cric_classes1)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
np.random.seed(0)
```

# As top 5 features seem to alternate on each run, iterating 10 times and taking the mean to find top 3 features

```python
from sklearn.ensemble import RandomForestClassifier
avg = []
for i in range(10):
    rfc = RandomForestClassifier()
    rfc.fit(cricX1, np.int8(cricY1))
    pred = rfc.predict(cricX1)
    print(accuracy_score(np.int8(cricY1),pred))
    importance_scores = rfc.feature_importances_
    feature_importances = [(feature, score) for feature, score in zip(cric_features1, importance_scores)]
    avg.append(sorted(feature_importances, key=lambda x: -x[1])[:5])

```

# Top Features on 10 iterations given by fea

```python
fea = []
for i in avg:
    for k in i:
        if k[0] not in fea:
            fea.append(k[0])
```

```python
fea
```

```python
im_fea={}
for j in fea:
    temp =[]
    for k in avg:
        for i in k:
            if i[0]==j:
                temp.append(i[1])
    im_fea[j]=np.array(temp).mean()
    
```

# TOP 3 FEATURES USING SIMPLE IMPUTER

```python
top3 = list(im_fea.keys())[:3]
top3
```

```python
# SORTING TOP 5 AND TAKING 10 ITERATIONS, SIMULATANEOUSLY WE CAN SEE THE ACCURACY OF RANDOM FOREST CLASSIFIER
from sklearn.metrics import confusion_matrix,accuracy_score
np.random.seed(0)
from sklearn.ensemble import RandomForestClassifier
avg1 = []
for i in range(10):
    rfc = RandomForestClassifier()
    rfc.fit(cricX2, np.int8(cricY2))
    pred = rfc.predict(cricX2)
    print(accuracy_score(np.int8(cricY2),pred))
    importance_scores = rfc.feature_importances_
    feature_importances = [(feature, score) for feature, score in zip(cric_features2, importance_scores)]
    avg1.append(sorted(feature_importances, key=lambda x: -x[1])[:5])
```

```python
fea1 = []
for i in avg1:
    for k in i:
        if k[0] not in fea1:
            fea1.append(k[0])
```
